````python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
````

````python
# 케라스로 구현하는 단순 선형회귀
x = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 공부 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 성적

model = Sequential()

# 출력 y의 차원은 1. 입력 x의 차원(input_dim)은 1
# 선형 회귀이므로 activation은 'linear'
model.add(Dense(1, input_dim=1, activation='linear'))

# sgd는 경사 하강법을 의미. 학습률(learning rate, lr)은 0.01.
sgd = optimizers.SGD(lr=0.01)

# 손실 함수(Loss function)은 평균제곱오차 mse를 사용합니다.
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])

model.fit(x, y, epochs=100, verbose=2)
````

````python
plt.plot(x, model.predict(x), 'r', x, y, 'k.')
````

### 기울기 소실, 폭주 문제 해결
1. 은닉층에 ReLU 사용  
<br>
2. Gradient Clipping : 기울기가 임계값을 넘지 못하게 자르기  
Adam = optimizers.Adam(lr=0.0001, clipnorm=1.)  
<br>
3. 가중치 초기화  
이전 층의 뉴런의 개수 $n_{in}$, 다음 층의 뉴런의 개수  $n_{out}$  
    * 세이비어 초기화 : 시그모이드나 탄젠트랑 같이 써야 좋음  

$$ W \sim Uniform(-\sqrt{\frac{6}{ n_{in} + n_{out} }}, +\sqrt{\frac{6}{ n_{in} + n_{out} }}) $$정규분포  $$ σ=\sqrt{\frac { 2 }{ n_{ in }+n_{ out } } } $$  

* He 초기화 : 렐루 계열이랑 잘 맞음  

$$ W\sim Uniform(- \sqrt{\frac { 6 }{ n_{in} } } , \space\space + \sqrt{\frac { 6 }{ n_{ in } } } ) $$  정규분포  $$ σ=\sqrt{\frac { 2 }{ n_{ in } } } $$  


4. 배치 정규화 : 활성화 함수를 통과하기 전 입력들에 대해 정규화 진행. 이후 스케일($\tau$)과 시프트($\beta$) 수행.  
Input : 미니 배치 $B = \{{x}^{(1)}, {x}^{(2)}, ..., {x}^{(m)}\}$  
Output : $y^{(i)} = BN_{γ, β}(x^{(i)})$  
$\hat{x}^{(i)} ← \frac{x^{(i)} - μ_{B}}{\sqrt{σ^2_{B}+ε}}$  정규화  
$y^{(i)} ← γ\hat{x}^{(i)} + β = BN_{γ, β}(x^{(i)})$ 스케일 조정(γ)과 시프트(β)를 통한 선형 연산

* 배치 정규화는 기울기 소실 문제 개선이 잘 되며, 가중치 초기화에 훨씬 덜 민감해지고 학습속도도 개선되는 등 장점 투성이지만,  배치 크기가 어느정도 되어야하며 RNN에서는 잘 안쓰인다.

## Subclassing API 배워보기
![image](https://wikidocs.net/images/page/106897/1_WzwKtnA0LEhiCGdWTTpLaA.png)

선형회귀
```python
import tensorflow as tf

class LinearRegression(tf.keras.Model):
  def __init__(self):
    super(LinearRegression, self).__init__()
    self.linear_layer = tf.keras.layers.Dense(1, input_dim=1, activation='linear')

  def call(self, x):
    y_pred = self.linear_layer(x)

    return y_pred

model = LinearRegression()

X = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 공부하는 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 각 공부하는 시간에 맵핑되는 성적

```

````python
train_email.head()
````

````python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
swords = set(stopwords.words("english"))

def changing(data):
    aa = data.apply(lambda x : re.sub("[^a-zA-Z]", ' ', x))
    aa = aa.apply(lambda x : x.split())
    new_data = []
    for text in aa:
        texts = [i for i in text if i not in swords]
        texts = " ".join(texts)
        new_data.append(texts)
    new_data = pd.Series(new_data)
    return new_data

trains = changing(train_email)

new_data = []
for i in test_email:
    aa = re.sub("[^a-zA-Z]", ' ', i)
    aa = aa.split()
    texts = [i for i in aa if i not in swords]
    texts = " ".join(texts)
    new_data.append(texts)
tests = pd.Series(new_data)
````

````python
vocab_size = 10000
num_classes = 20

def prepare_data(train_data, test_data, mode): # 전처리 함수
    tokenizer = Tokenizer(num_words = vocab_size) # vocab_size 개수만큼의 단어만 사용한다.
    tokenizer.fit_on_texts(train_data) 
    X_train = tokenizer.texts_to_matrix(train_data, mode=mode) # 샘플 수 × vocab_size 크기의 행렬 생성
    X_test = tokenizer.texts_to_matrix(test_data, mode=mode) # 샘플 수 × vocab_size 크기의 행렬 생성
    return X_train, X_test, tokenizer.index_word

X_train, X_test, index_to_word = prepare_data(trains, tests, 'binary') # binary 모드로 변환
y_train = to_categorical(train_label, num_classes) # 원-핫 인코딩
y_test = to_categorical(test_label, num_classes) # 원-핫 인코딩
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
````

````python
def fit_and_evaluate(X_train, y_train, X_test, y_test):
    model = Sequential()
    model.add(Dense(256, input_shape=(vocab_size,), activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X_train, y_train, batch_size=128, epochs=5, verbose=1, validation_split=0.1)
    score = model.evaluate(X_test, y_test, batch_size=128, verbose=0)
    return score[1]
````

````python
modes = ['binary', 'count', 'tfidf', 'freq'] # 4개의 모드를 리스트에 저장.

for mode in modes: # 4개의 모드에 대해서 각각 아래의 작업을 반복한다.
    X_train, X_test, _ = prepare_data(trains, tests, mode) # 모드에 따라서 데이터를 전처리
    score = fit_and_evaluate(X_train, y_train, X_test, y_test) # 모델을 훈련하고 평가.
    print(mode+' 모드의 테스트 정확도:', score)
````

````python
# super() : 부모 클래스에서 도움받기
class person():
    def __init__(self, name):
        self.name = name

class person2(person):
    def __init__(self, name, email):
        super().__init__(name)
        self.email = email
````

````python
class hunt():
    def __init__(self, person, animal):
        self.person = person
        self.animal = animal
    def who(self):
        return self.person
    def what(self):
        return self.animal
````

````python
hunts = hunt('kim','pig')
print(hunts.who(), ':', hunts.what(),'!')
````

````python
class calculator():
    def __init__(self):
        pass
    
    def ans(self,num1,num2):
        return '%d + %d ='%(num1, num2)
    def plus(self,num1,num2):
        return num1 + num2
````

````python
cal = calculator()
print(cal.ans(1,2), cal.plus(1,2))
````
